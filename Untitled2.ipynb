```python
def evaluate(U, I, threshold=4, N=1000, print_PR=True):
    """
    threshold: レコメンドが正解だったと言える評価値の下限。デフォルトでは4以上の評価ならレコメンドして正解とした。
    N: 評価予測値の上位N個に関してモデルを評価する。
    """
    pred = np.dot(U, I.T)

    # validation
    data = []
    with open('ml-100k/u1.test') as f:
        for line in f:
            uid, iid, rate, _ = map(int, line.split())
            data.append((uid-1, iid-1, rate, pred[uid-1, iid-1]))
    data.sort(key=lambda x: x[3], reverse=True)
    
    TruePositive = 0
    FalsePositive = 0
    for i in range(N):
        if data[i][2] >= threshold:
            TruePositive += 1
        else:
            FalsePositive += 1
    Precision = TruePositive / N
    Recall = TruePositive / len([d for d in data if d[2] >= threshold])
    
    if print_PR:
        print('Precision: {}'.format(Precision))
        print('Recall: {}'.format(Recall))
        print('F1-measure: {}'.format(2 * Precision * Recall / (Precision + Recall)))
    
    return Precision, Recall
```

```python
k = 20
user = 943
item = 1682

eps = 0.5
alpha = 1e-2
lambda_u = 1e-1
lambda_i = 1e-1
epoch = 1000

data = []
R = np.zeros((user, item))
U = np.random.random((user, k))
I = np.random.random((item, k))

with open('ml-100k/u1.base') as f:
    for line in f:
        uid, iid, rate, _ = map(int, line.split())
        data.append((uid-1, iid-1, rate))
        R[uid-1, iid-1] = rate
M = R.astype(np.bool)

# training
for i in range(epoch):
    for uid, iid, rate in data:
        error = rate - U[uid] @ I[iid]
        #error = np.sign(error)*eps**2 if abs(error) < eps else error
        #error = 0 if abs(error) < eps else np.sign(error)
        delta_U = alpha * (error * I[iid] - lambda_u * U[uid])
        delta_I = alpha * (error * U[uid] - lambda_i * I[iid])
        U[uid] += delta_U
        I[iid] += delta_I
    
    E = R - U @ I.T
    rmse = np.sqrt(np.mean((E*M)**2))
    print(i, rmse)
```

### ALSと比較

```python
k = 100
user = 943
item = 1682

eps = 0.5
alpha = 1e-2
lambda_u = 1e-1
lambda_i = 1e-1
epoch = 100

data = []
R = np.zeros((user, item))
U = np.random.random((user, k))
I = np.random.random((item, k))

with open('ml-100k/u1.base') as f:
    for line in f:
        uid, iid, rate, _ = map(int, line.split())
        data.append((uid-1, iid-1, rate))
        R[uid-1, iid-1] = rate
M = R.astype(np.bool)

# training
for i in range(epoch):
    random.shuffle(data)
    for uid, iid, rate in data:
        error = rate - U[uid] @ I[iid]
        delta_U = alpha * (error * I[iid] - lambda_u * U[uid])
        delta_I = alpha * (error * U[uid] - lambda_i * I[iid])
        U[uid] += delta_U
        I[iid] += delta_I
    
    if (i + 1) % 10 == 0:
        E = R - U @ I.T
        rmse = np.sqrt(np.mean((E*M)**2))
        print(i+1, rmse)
```

```python
P, R = evaluate(U, I, threshold=4, N=1, print_PR=False)
Ps, Rs = [P], [R]
AUC = 0
for i in range(100, 20001, 100):
    P, R = evaluate(U, I, threshold=4, N=i, print_PR=False)
    Ps.append(P)
    Rs.append(R)
    AUC += (Ps[-1] + Ps[-2]) * (Rs[-1] - Rs[-2]) / 2
plt.plot(Rs, Ps)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
print('AUC = {}'.format(AUC))
```
