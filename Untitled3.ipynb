```python
k = 100
user = 943
item = 1682

lambda_u = 1e+1
lambda_i = 1e+1
epoch = 10

data = []
R = np.zeros((user, item), dtype=np.int8)
U = np.zeros((user, k))
I = np.random.random((item, k))
CU = np.zeros((user, item, item), dtype=np.int8)
CI = np.zeros((item, user, user), dtype=np.int8)


with open('ml-100k/u1.base') as f:
    for line in f:
        uid, iid, rate, _ = map(int, line.split())
        data.append((uid-1, iid-1, rate))
        R[uid-1, iid-1] = rate
        CU[uid-1, iid-1, iid-1] = 1
        CI[iid-1, uid-1, uid-1] = 1
lambdaU = lambda_u * np.eye(k)
lambdaI = lambda_i * np.eye(k)
for iid in range(item):
    count = R[:, iid].clip(0,1).sum()
    if count > 0:
        I[iid, 0] = R[:, iid].sum() / count

# training
print('training start.')
for i in range(epoch):
    # optimize U
    for uid in range(user):
        U[uid] = R[uid] @ I @ np.linalg.inv(I.T @ CU[uid] @ I + lambdaU)
    
    # optimize I
    for iid in range(item):
        I[iid] = R[:, iid] @ U @ np.linalg.inv(U.T @ CI[iid] @ U + lambdaI)

    E = R - U @ I.T
    rmse = np.sqrt(np.mean((E[R>0])**2))
    print(i+1, rmse)
```

評価にはRMSE@TopNを使うか、
閾値を設けてレコメンドするべきものか否かを判定

```python
def evaluate(U, I, threshold=4, N=1000, print_PR=True):
    """
    threshold: レコメンドが正解だったと言える評価値の下限。デフォルトでは4以上の評価ならレコメンドして正解とした。
    N: 評価予測値の上位N個に関してモデルを評価する。
    """
    pred = np.dot(U, I.T)

    # validation
    data = []
    with open('ml-100k/u1.test') as f:
        for line in f:
            uid, iid, rate, _ = map(int, line.split())
            data.append((uid-1, iid-1, rate, pred[uid-1, iid-1]))
    data.sort(key=lambda x: x[3], reverse=True)
    
    TruePositive = 0
    FalsePositive = 0
    for i in range(N):
        if data[i][2] >= threshold:
            TruePositive += 1
        else:
            FalsePositive += 1
    Precision = TruePositive / N
    Recall = TruePositive / len([d for d in data if d[2] >= threshold])
    
    if print_PR:
        print('Precision: {}'.format(Precision))
        print('Recall: {}'.format(Recall))
        print('F1-measure: {}'.format(2 * Precision * Recall / (Precision + Recall)))
    
    return Precision, Recall
```

```python
Ps, Rs = [], []
for i in range(100, 20001, 1000):
    P, R = evaluate(U, I, threshold=4, N=i, print_PR=False)
    Ps.append(P)
    Rs.append(R)
plt.plot(Rs, Ps)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```

```python
k = 100
user = 943
item = 1682

lambda_u = 1e+1
lambda_i = 1e+1
epoch = 10

data = []
R = np.zeros((user, item), dtype=np.int8)
U = np.zeros((user, k))
I = np.random.random((item, k))
CU = np.zeros((user, item, item), dtype=np.int8)
CI = np.zeros((item, user, user), dtype=np.int8)


with open('ml-100k/u1.base') as f:
    for line in f:
        uid, iid, rate, _ = map(int, line.split())
        data.append((uid-1, iid-1, rate))
        R[uid-1, iid-1] = rate
        CU[uid-1, iid-1, iid-1] = 1
        CI[iid-1, uid-1, uid-1] = 1
lambdaU = lambda_u * np.eye(k)
lambdaI = lambda_i * np.eye(k)
for iid in range(item):
    count = R[:, iid].clip(0,1).sum()
    if count > 0:
        I[iid, 0] = R[:, iid].sum() / count

# training
print('training start.')
for i in range(epoch):
    # optimize U
    for uid in range(user):
        U[uid] = R[uid] @ I @ np.linalg.inv(I.T @ CU[uid] @ I + lambdaU)
    
    # optimize I
    for iid in range(item):
        I[iid] = R[:, iid] @ U @ np.linalg.inv(U.T @ CI[iid] @ U + lambdaI)

    E = R - U @ I.T
    rmse = np.sqrt(np.mean((E[R>0])**2))
    print(i+1, rmse)
```

```python
P, R = evaluate(U, I, threshold=4, N=1, print_PR=False)
Ps, Rs = [P], [R]
AUC = 0
for i in range(100, 20001, 100):
    P, R = evaluate(U, I, threshold=4, N=i, print_PR=False)
    Ps.append(P)
    Rs.append(R)
    AUC += (Ps[-1] + Ps[-2]) * (Rs[-1] - Rs[-2]) / 2
plt.plot(Rs, Ps)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
print('AUC = {}'.format(AUC))
```
