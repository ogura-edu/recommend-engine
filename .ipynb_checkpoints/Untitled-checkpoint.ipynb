# Alternating Least Square

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 5
user = 943
item = 1682
rating = 100000

threshold = 1e-3
alpha = 1e-3
epoch = 3000
```

```python
A = np.zeros((user, item))
X = np.random.random((user, k))
Y = np.random.random((item, k))
```

```python
with open('ml-100k/u1.base') as f:
    for line in f:
        uid, iid, rat, _ = map(int, line.split())
        A[uid-1, iid-1] = rat
B = A.astype(np.bool) # ratingがある部分はどこか。マスキング用
```

```python
# training

for i in range(epoch):
    # update X
    E = np.dot(X, Y.T)*B - A
    delta_X = np.dot(E, Y)
    X -= alpha * delta_X
    
    # update Y
    E = np.dot(X, Y.T)*B - A
    delta_Y = np.dot(E.T, X)
    Y -= alpha * delta_Y
    
    E = np.dot(X, Y.T)*B - A
    error = np.mean(E**2)
    print(error)
```

### 評価値をclipしてみると…

#### k = 5

```python
pred = np.dot(X, Y.T).clip(1, 5)

# validation
sub_val = []
with open('ml-100k/u1.base') as f:
    for line in f:
        uid ,iid, rat, _ = map(int, line.split())
        #print(uid, iid, rat, pred[uid-1, iid-1])
        
        # 評価値と予測値がかけ離れていないかcheck
        sub_val.append(pred[uid-1, iid-1] - rat)

plt.hist(sub_val, bins=100, range=(-4, 4))
plt.show()

print('特徴量：{}'.format(k))
print(np.std(sub_val))
```

```python
# test

sub_test = []
with open('ml-100k/u1.test') as f:
    for line in f:
        uid, iid, rat, _ = map(int, line.split())
        #print(uid, iid, rat, pred[uid-1, iid-1])
        sub_test.append(pred[uid-1, iid-1] - rat)

plt.hist(sub_test, bins=100, range=(-4, 4))
plt.show()

print('特徴量：{}'.format(k))
print(np.std(sub_test))
```

### ε-感度損失を使ってみる

```python
k = 5
user = 943
item = 1682

eps = 0.5
alpha = 1e-3
epoch = 1000

A = np.zeros((user, item))
X = np.random.random((user, k))
Y = np.random.random((item, k))

with open('ml-100k/u1.base') as f:
    for line in f:
        uid, iid, rat, _ = map(int, line.split())
        A[uid-1, iid-1] = rat
B = A.astype(np.bool) # ratingがある部分はどこか。マスキング用

# training
for i in range(epoch):
    # update X
    E = np.dot(X, Y.T)*B - A
    dL_dE = np.where(abs(E) < eps, np.sign(E) * eps**2, E)
    delta_X = np.dot(dL_dE, Y)
    X -= alpha * delta_X
    
    # update Y
    E = np.dot(X, Y.T)*B - A
    dL_dE = np.where(abs(E) < eps, np.sign(E) * eps**2, E)
    delta_Y = np.dot(dL_dE.T, X)
    Y -= alpha * delta_Y
    
    if (i+1) % 100 == 0:
        E = np.dot(X, Y.T)*B - A
        error = np.sqrt(np.mean(E**2))
        print(i+1, error)
```

### バッチ最適化してみる

```python
import random

k = 10
user = 943
item = 1682

eps = 0.5
alpha = 1e-3
epoch = 1000
batch_size = 10000

X = np.random.random((user, k))
Y = np.random.random((item, k))

data = []
with open('ml-100k/u1.base') as f:
    for line in f:
        uid, iid, rat, _ = map(int, line.split())
        data.append((uid-1, iid-1, rat))

# training
for i in range(epoch):
    batch = random.sample(data, batch_size)
    A = np.zeros((user, item))
    for uid, iid, rat in batch:
        A[uid, iid] = rat
    B = A.astype(np.bool) # ratingがある部分はどこか。マスキング用
    
    # update X
    E = np.dot(X, Y.T)*B - A
    dL_dE = np.where(abs(E) < eps, np.sign(E) * eps**2, E)
    delta_X = np.dot(dL_dE, Y)
    X -= alpha * delta_X
    
    # update Y
    E = np.dot(X, Y.T)*B - A
    dL_dE = np.where(abs(E) < eps, np.sign(E) * eps**2, E)
    delta_Y = np.dot(dL_dE.T, X)
    Y -= alpha * delta_Y
    
    if (i+1) % 100 == 0:
        E = np.dot(X, Y.T)*B - A
        error = np.sqrt(np.mean(E**2))
        print(i+1, error)
```
